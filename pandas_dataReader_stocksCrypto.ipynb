## FRED - Federal Reserve Economic Data

```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2010, 1, 1)

end = datetime.date.today()
```

**GDP Data**

```python
gdp_fred = web.DataReader('GDP', 'fred', start, end)
```

```python
gdp_fred.tail(5)
```

```python
# Simple Plot of FRED GDP Values:

gdp_fred['GDP'].plot(title = 'FRED GDP Values\n', grid = True, figsize = (15, 5))
plt.xlabel('\n Year')
plt.ylabel('Value \n')
plt.show()
```

Dip in GDP is from COVID 19 lockdowns around early 2020.


## NASDAQ Data
