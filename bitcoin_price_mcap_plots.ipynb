```python
# Import packages:

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Bitcoin historical price from coincodex.com
# Date from Aug 16, 2010 to April 3, 2022 Setting

btc_df = pd.read_csv('bitcoin_2010-8-16_2022-4-3.csv')
```

```python
# Preview
btc_df.head(10)
```

```python
# Preview
btc_df.tail(10)
```

```python
# Set date as datetime:
btc_df['Date'] = pd.to_datetime(btc_df['Date'])
```

```python
# Change date order

btc_df2 = btc_df.sort_values(by = 'Date')
```

```python
btc_df2
```

## Line Plot Of Bitcoin Price With Plotly

```python
fig = px.line(btc_df2, x='Date', y="Close")

fig.update_layout(
    title={
        'text': 'Bitcoin Price From Aug 16, 2010 to Apr 3, 2022',
        'y':0.95,&n
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font_color = '#60974A'
)

# Save as png static image, cant do interactivity in github.
fig.show("png")

```

```python
# Another way

fig = go.Figure(data=go.Scatter(x=btc_df2['Date'], 
                                y= btc_df2['Close']))

fig.update_layout(
    title={
        'text': 'Bitcoin Price From Aug 16, 2010 to Apr 3, 2022',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}
)
fig.show("png")
```

### Marketcap of Bitcoin

```python
fig = px.line(btc_df2, x='Date', y="Market Cap")

fig.update_layout(
    title={
        'text': 'Bitcoin Market Cap From Aug 16, 2010 to Apr 3, 2022',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    title_font_color = 'blue'
)
fig.show("png")
```
